```
from sklearn.cluster import KMeans
from sklearn import datasets
iris=datasets.load_iris()
X = iris.data[:]

kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
print(kmeans.cluster_centers_)

plt.scatter(X[:,2],X[:,3], c=kmeans.labels_, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,2] ,kmeans.cluster_centers_[:,3], color='black', marker='*',label='centroids' )
plt.legend(scatterpoints=1)
plt.show()
```

```
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering

# create dendrogram
dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
```

```
# create clusters
cluster = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')
cluster.fit_predict(X)
plt.scatter(X[:,2],X[:,3], c=cluster.labels_, cmap='rainbow')
```

```
#Question 3 : Inplementing K_means


import math
from  matplotlib import pyplot as plt
sns.set(style='ticks')
np.random.seed(0)
N = 37
count_it = 0
from sklearn.datasets import load_iris
iris = load_iris()
data1 = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                     columns= iris['feature_names'] + ['target'])
j_values = []
x_values = []
z_values = []
j = 0
hi = 0
c1 = data1.iloc[0]
c2 = data1.iloc[34]
c3 = data1.iloc[110]
count = 1
new_c1 = [0,0,0,0]
new_c2 = [0,0,0,0]
new_c3 = [0,0,0,0]
while hi!=12 and count_it < 20:
  count_it += 1
  j = 0
  hi = 0
  count1 = 0
  count2 = 0
  count3 = 0
  if count != 1:
    c1[0] = new_c1[0]
    c1[1] = new_c1[1]
    c1[2] = new_c1[2]
    c1[3] = new_c1[3]
    c2[0] = new_c2[0]
    c2[1] = new_c2[1]
    c2[2] = new_c2[2]
    c2[3] = new_c2[3]
    c3[0] = new_c3[0]
    c3[1] = new_c3[1]
    c3[2] = new_c3[2]
    c3[3] = new_c3[3]
  for x in range(len(data1)):
    current_node = data1.iloc[x]
    distance1 = math.sqrt(pow(current_node[0]-c1[0],2) + pow(current_node[1]-c1[1],2) + pow(current_node[2]-c1[2],2) + pow(current_node[3]-c1[3],2))
    distance2 = math.sqrt(pow(current_node[0]-c2[0],2) + pow(current_node[1]-c2[1],2) + pow(current_node[2]-c2[2],2) + pow(current_node[3]-c2[3],2))
    distance3 = math.sqrt(pow(current_node[0]-c3[0],2) + pow(current_node[1]-c3[1],2) + pow(current_node[2]-c3[2],2) + pow(current_node[3]-c3[3],2))
    final_distance = min(distance1,distance2,distance3)
    if final_distance == distance1:
      current_node[4] = 1
    if final_distance == distance2:
      current_node[4] = 2
    if final_distance == distance3:
      current_node[4] = 3
  new_c1 = [0,0,0,0]
  new_c2 = [0,0,0,0]
  new_c3 = [0,0,0,0]
  for x in range(len(data1)):
    current = data1.iloc[x]
    if current[4] == 1:
      count1 = count1 + 1
      new_c1 = [new_c1[0] + current[0],new_c1[1] + current[1],new_c1[2] + current[2],new_c1[3] + current[3]]
    if current[4] == 2:
      count2 = count2 + 1
      new_c2 = [new_c2[0] + current[0],new_c2[1] + current[1],new_c2[2] + current[2],new_c2[3] + current[3]]
    if current[4] == 3:
      count3 = count3 + 1
      new_c3 = [new_c3[0] + current[0],new_c3[1] + current[1],new_c3[2] + current[2],new_c3[3] + current[3]]
  new_c1[:] = [x / count1 for x in new_c1]
  new_c2[:] = [x / count2 for x in new_c2]
  new_c3[:] = [x / count3 for x in new_c3]
  count = 2
  fg = seaborn.FacetGrid(data=data1, hue='target')
  fg.map(plt.scatter, 'sepal length (cm)', 'sepal width (cm)').add_legend()
  if new_c1[0] == c1[0]:
    hi = hi + 1
  if new_c1[1] == c1[1]:
    hi = hi + 1
  if new_c1[2] == c1[2]:
    hi = hi + 1
  if new_c1[3] == c1[3]:
    hi = hi + 1
  if new_c2[0] == c2[0]:
    hi = hi + 1
  if new_c2[1] == c2[1]:
    hi = hi + 1
  if new_c2[2] == c2[2]:
    hi = hi + 1
  if new_c2[3] == c2[3]:
    hi = hi + 1
  if new_c3[0] == c3[0]:
    hi = hi + 1
  if new_c3[1] == c3[1]:
    hi = hi + 1
  if new_c3[2] == c3[2]:
    hi = hi + 1
  if new_c3[3] == c3[3]:
    hi = hi + 1
  for x in range(len(data1)):
    current = data1.iloc[x]
    if current[4] == 1:
      j = j + pow(current_node[0]-c1[0],2) + pow(current_node[1]-c1[1],2) + pow(current_node[2]-c1[2],2) + pow(current_node[3]-c1[3],2)
    if current[4] == 2:
      j = j + pow(current_node[0]-c2[0],2) + pow(current_node[1]-c2[1],2) + pow(current_node[2]-c2[2],2) + pow(current_node[3]-c2[3],2)
    if current[4] == 3:
      j = j + pow(current_node[0]-c3[0],2) + pow(current_node[1]-c3[1],2) + pow(current_node[2]-c3[2],2) + pow(current_node[3]-c3[3],2)
  j_values.append(j)
print(j_values)
for x in range(1,len(j_values)+1):
  x_values.append(x)
for x in j_values:
  z_values.append(x)
plt.plot(x_values,z_values)
plt.xlabel('x - axis') 
plt.ylabel('y - axis')
```
